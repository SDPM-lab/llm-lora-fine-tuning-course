```python
import numpy as np
import matplotlib.pyplot as plt

# 建立 x 值，避免 x=0 因為 log(0) 未定義
x = np.linspace(0.001, 1, 1000)  # 從 0.001 開始到 1
y = np.log2(x)

# 繪製圖形
plt.figure(figsize=(6, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = log(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) = log(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(0, 1)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.show()

# 顯示一些關鍵點的數值
print(f"log(1.0) = {np.log2(1.0):.2f}")
print(f"log(0.5) = {np.log2(0.5):.2f}")
print(f"log(0.1) = {np.log2(0.1):.2f}")
```

```python
# Entropy of the distribution

def entropy(x: list[float]) -> list[float]:
    result = -x * np.log2(x)
    result += -(1 - x) * np.log2(1 - x)
    return result

plt.figure(figsize=(6, 6))
plt.plot(x, entropy(x), 'b-', linewidth=2, label='Entropy H(x)')
plt.xlabel('x')
plt.ylabel('H(x)')
plt.title('Entropy')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(0, 1)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.show()
```

```python
print(entropy(np.array([0.99])[0]))
print(entropy(np.array([0.4])[0]))
```

```python
# −(𝟏∗𝒍𝒐𝒈𝟐(𝟎.𝟔)+𝟎∗𝒍𝒐𝒈𝟐(𝟎.𝟒))
result = -(1 * np.log2(0.6) + 0 * np.log2(0.4))
print(f"Entropy result: {result:.3f}")

# 𝟎.𝟕𝟑𝟕=−(𝟎∗𝒍𝒐𝒈𝟐(𝟎.𝟐)+𝟏∗𝒍𝒐𝒈𝟐(𝟎.𝟖))
result = -(0 * np.log2(0.2) + 1 * np.log2(0.8))
print(f"Entropy result: {result:.3f}")
```
